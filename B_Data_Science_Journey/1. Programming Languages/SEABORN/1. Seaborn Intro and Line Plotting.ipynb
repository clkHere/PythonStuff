<h1> Seaborn Line Plotting</h1> 

Line plots are graphs that show <b>frequency of data along a number line</b>. Best practice is to use line plots when comparing fewer than 25 numbers as they are a quick and simple way to organize data. 


<hr/>

<h2>Importing Seaborn and Matplotlib</h2>

Seaborn works on top of <b>matplotlib</b> - which must be imported to use Seaborn. 

The two are imported as the following: 

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

Seaborn has a built in dataset called 'fmri' which can be used to test out Seaborn's features. To use it, a simple call using the <code>sns.load_dataset('fmri')</code> is used:

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

<hr/>

<h2>Line Plotting</h2>

To make a line plot in Seaborn, we use the <code>sns.lineplot()</code> method. 

The <code>sns.lineplot()</code> method takes three primary parameters:

<ul> <code>x</code> = COL_NAME -- The first parameter</ul>
<ul> <code>y</code> = COL_NAME -- The second parameter</ul>
<ul> <code>data</code> = dataset name -- The last parameter</ul>

Example: 

```python
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

<b> Analysis </b>

Looking at the lineplot, we can see that the 'signal' increases between 'timepoint' 1.0 to 5.0, then decreases from 5.0 to 11.0. 

<hr/>

<h2>Adding a Comparison with <code>hue='COL_NAME'</code></h2>

The <code>hue=</code> parameter is an additional parameter that overlays another column on the line plot:

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event')
plt.show()
```

<b> Analysis </b> 

We see from the added 'event' hue that values 'stim' and 'cue' were plotted on the map from the 'event' column. From those events, we see that the 'stim' has a larger time/signal rise/drop/stablize event than the cue. Conversely, the 'cue' event takes less time to stablize. 

<hr/>

<h2>Adding Styles with the <code>style</code> Parameter</h2>

To add a style, simply add the <code>style=</code> parameter and the column that you wish to apply the style to:

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event')
plt.show()
```

<hr/>

<h2>Adding Markers with the <code>markers=</code> Parameter</h2>

Similar to adding a style, simply add the <code>markers=True</code> parameter to the plot code and Seaborn will automatically set visual markers at each change in data values: 

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event', markers=True)
plt.show()
```
